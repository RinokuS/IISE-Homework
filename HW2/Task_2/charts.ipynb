# Генерация графиков

Для точности измерений были предприняты следующие меры:
* до начала измерений было сделано 2^30 инкрементов
* измерения для каждого размера проводились 10 раз, результат усреднился

### График времени доступа вокруг L1 кэша

```python
fig, axes = plt.subplots(1)

axes.plot(data_l1['Size'], data_l1['Timings'])
axes.set_title('L1 cache check')
axes.set_xlabel('array of 64-byte structures size')
axes.set_ylabel('time of 2^27 increments (seconds)')
fig.set_figheight(10)
fig.set_figwidth(15)
```

### График времени доступа вокруг L2 кэша

```python
fig, axes = plt.subplots(1)

axes.plot(data_l2['Size'], data_l2['Timings'])
axes.set_title('L2 cache check')
axes.set_xlabel('array of 64-byte structures size')
axes.set_ylabel('time of 2^27 increments (seconds)')
fig.set_figheight(10)
fig.set_figwidth(15)
```
