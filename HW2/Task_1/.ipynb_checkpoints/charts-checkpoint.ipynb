# Генерация графиков

```python
types_of_programs = ['1 thread', '5 threads', '10 threads', '20 threads']
```

Измерения проводитились для удобства на квадратных матрицах разных размерностей с 10 эпохами жизни.
Для генерации матриц большой размерности в программу была добавлена логика генерации матриц без мануального ввода, а для обепечения идентичной генерации при профилировании задан константный сид рандомизации (также идентичность генерации была проверена diffchecker'ом с помощью выводов сгенерированных матриц в файл).

P.S. Для упрощения измерения из программы на момент измерения убирались все выводы, ибо они могли занимать много времени, поэтому программы на матрице 15000х15000 работают в среднем в 2 раза быстрее, чем должны были бы, это не ошибка.

### График времени работы программы

```python
fig, axes = plt.subplots(1)

for t in types_of_programs:
    axes.plot(data['Size'], data[t], label=t)
axes.legend()
axes.set_title('10 epochs')
axes.set_xlabel('size of square game grid')
axes.set_ylabel('time (nanoseconds)')
fig.set_figheight(15)
fig.set_figwidth(15)
```
